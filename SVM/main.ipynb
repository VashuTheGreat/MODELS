```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("deforestation_dataset.csv")
```

```python
data.head()
```

```python
data['Deforestation_level']=data['Tree_Cover_Loss_percent']*(data['Forest_Loss_Area_km2']+data['Protected_Areas_Percent'])
```

```python
data.select_dtypes(include=['float64', 'int64']).corr()['Deforestation_level']
```

```python
features=data[['Forest_Loss_Area_km2','Tree_Cover_Loss_percent']]
x=np.array(features)
```

```python
y=data['Deforestation_level']
y=np.array(y)
```

```python
x
```

```python
y.reshape(-1,1)
```

```python
from sklearn import svm
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
y_scaled=scaler.fit_transform(y.reshape(-1,1))
```

```python
x_scaled
```

```python
y_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR

svr_linear=SVR(kernel='linear',C=1.0,epsilon=0.1)
svr_linear.fit(x_train, y_train)
```

```python
y_pred_scaled = svr_linear.predict(x_test)
```

```python
y_pred=scaler.inverse_transform(y_pred_scaled.reshape(-1,1)).flatten()
```

```python
y_pred
```

```python
y_test
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("SVR Linear Model Performance:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")
```
