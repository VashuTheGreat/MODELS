```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Global_Pollution_Analysis.csv')
```

```python
data
```

```python
data = data.drop(columns=['Country', 'Year'])
```

```python
data
```

```python
data.corr()['Energy_Recovered (in GWh)']
```

```python
data_input=data['Industrial_Waste (in tons)']
data_inputy=data['Energy_Recovered (in GWh)']
```

```python
x=np.array(data_input)
y=np.array(data_inputy)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python

import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split



model = keras.Sequential([
    keras.layers.Dense(8, activation='relu', input_shape=(1,)),  # FIXED: layers.Dense
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1)  # Output Layer
])
model.compile(optimizer='adam',loss='mse')
model.fit(x,y,epochs=100,verbose=0)
```

```python
new_waste = np.array([[600]])  # Example: 600 tons of waste
predicted_energy = model.predict(new_waste)
print("Predicted Energy Recovered:", predicted_energy[0][0], "GWh")
```
