```python
total_pollution_score=(df['Air_Pollution_Index']+df['Water_Pollution_Index']+df['Soil_Pollution_Index'])/3

```

```python
total_pollution_score=np.array(total_pollution_score)
```

```python
condition = np.empty(len(total_pollution_score), dtype=object)
```

```python
def rate(total_pollution_score, condition):
    for i in range(len(total_pollution_score)):
        if 0 <= total_pollution_score[i] < 51:
            condition[i] = "Good"
        elif 51 <= total_pollution_score[i] < 101:
            condition[i] = "Moderate"
        else:
            condition[i] = "High"
```

```python
rate(total_pollution_score,condition)
```

```python
condition
```

```python
xl_train,xl_test,yl_train,yl_test=train_test_split(total_pollution_score,condition,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

modelL=LogisticRegression()
modelL.fit(xl_train.reshape(-1,1),yl_train)
```

```python
yl_pred = modelL.predict(xl_test.reshape(-1,1))
```

```python
yl_pred
```

```python
yl_test
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(yl_test, yl_pred)
print(f"Accuracy: {accuracy:.4f}")

print("Confusion Matrix:")
print(confusion_matrix(yl_test, yl_pred))

print("Classification Report:")
print(classification_report(yl_test, yl_pred))
```
